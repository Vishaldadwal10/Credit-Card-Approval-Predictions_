Handling the missing values - checking for null values in data set

```python

import pandas as pd

# Load your dataset
file_path = '/Users/vishaldadwal/Downloads/crx.data'
df = pd.read_csv(file_path, header=None)

# Separate numeric and non-numeric columns
numeric_cols = df.select_dtypes(include='number').columns
non_numeric_cols = df.select_dtypes(exclude='number').columns

# Fill NaN values for numeric columns with the mean of the column
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Fill NaN values for non-numeric columns with a placeholder (like 'Unknown') or drop them
df[non_numeric_cols] = df[non_numeric_cols].fillna('Unknown')

# Verify if there are any NaN values left
print(df.isnull().sum())

```

Preprocessing the data (part 1)

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()


for col in df.columns:
    if df[col].dtype == 'object':  
        df[col] = le.fit_transform(df[col])  
```

splitting the set into train and test sets 

```python
from sklearn.model_selection import train_test_split
```

```python
df = df.drop([11, 13], axis =1)
df = df.to_numpy()
```

```python
X = df[:,0:13]
y = df[:,13]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                y,
                                test_size=0.33,
                                random_state=42)
```

Preprocessing the data - scaling 

```python
from sklearn.preprocessing import MinMaxScaler as MMS

scaler = MMS(feature_range=(0, 1))
```

```python
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.transform(X_test)
```

fitting logistic regression model 

```python
from sklearn.linear_model import LogisticRegression as logreg

logreg.fit(rescaledX_train, y_train)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import mutual_info_classif

df = pd.DataFrame(df)

# Calculate correlation matrix
correlation_matrix = np.corrcoef(df.T)  


X = df.iloc[:, :-1]  # Features
y = df.iloc[:, -1]   # Target
mutual_info = mutual_info_classif(X, y, discrete_features='auto')


mi_df = pd.DataFrame({'Feature': X.columns, 'Mutual Information': mutual_info})
mi_df = mi_df.sort_values(by='Mutual Information', ascending=False)


fig, ax = plt.subplots(1, 2, figsize=(20, 8))

sns.heatmap(correlation_matrix, annot=True, cmap='magma', ax=ax[0])
ax[0].set_title('Correlation Matrix')


sns.barplot(x='Mutual Information', y='Feature', data=mi_df, ax=ax[1])
ax[1].set_title('Mutual Information')
ax[1].tick_params(axis ='x', rotation = 30)

plt.suptitle('Correlation and Mutual Information Before Adding New Feature')
plt.tight_layout()
plt.show()

```
